```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_sqft(x):
    snippets = x.split("-")
    if(len(snippets) == 2):
        return ((float(snippets[0]) + float(snippets[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
#function to eliminate the outliers from this price_per_sqft 
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key , subdf in df.groupby("location"):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[((subdf.price_per_sqft) > (m - st)) & ((subdf.price_per_sqft) <= (m + st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index = True)
    return df_out
df3 = remove_pps_outliers(df2)
df3.shape
```

```python
# Function to check the total price distribution of a specific place of 2 and 3 bedroom flats 
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bedroom==2)]
    bhk3 = df[(df.location==location) & (df.bedroom==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    myplot.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    myplot.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    myplot.xlabel("Total Square Feet Area")
    myplot.ylabel("Price (Lakh Indian Rupees)")
    myplot.title(location)
    myplot.legend()
plot_scatter_chart(df3,'Rajaji Nagar')   
```

```python
#just remove the outliers.
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bedroom, bhk_df in location_df.groupby('bedroom'):
            bhk_stats[bedroom] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bedroom, bhk_df in location_df.groupby('bedroom'):
            stats = bhk_stats.get(bedroom-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df4 = remove_bhk_outliers(df3)
```

```python
df5 = df4[~(df4.bath > df4.bedroom + 2)]
```

```python
df5.shape
```

```python
df5 = df5.drop(['price_per_sqft'],axis = 'columns')
df5
```

```python
df6 = pd.get_dummies(df5['location'])
df6
```

```python
df7 = pd.concat([df5.drop(['location'],axis = 'columns'),df6],axis = 'columns')
df7
```

```python
# We will Build the Model now.
X = df7.drop(['price'],axis = 'columns')
X
```

```python
Y = df7.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
# We will Measure the Accuracy of our Linear Regression Model by K Fold Cross Validation
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=6, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
#As the cross validation scores are above 80% in all cases ,so we can deduce that our model is pretty good.
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
